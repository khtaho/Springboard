## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
#Load the world bank project file to be processed
data2=json.load((open('data/world_bank_projects.json')))
```

```python
#create DataFrame from nested elements with desired columns using json_normalized
df = pd.DataFrame.from_dict(json_normalize(data2,'mjtheme_namecode',['project_name','countryname']))
df.columns = ['code', 'theme', 'project_name', 'countryname']
df = df.sort_values(by=['countryname']) 
type(df)
df
```

```python
#Question 1, the countries with the most projects

#This creates the pivot table so we can see the top 10 countries with most projects
project_count=df.pivot_table(values='project_name',index='countryname',aggfunc='count')
#Sort by the values in an ascending order
project_count.sort_values(by=['project_name'], inplace=True,ascending=False)
#Rename the Index column name
project_count.index.names=['Country Name']
#Rename of Count column
project_count.columns=['Number of Projects']
#Print the top 10
print(project_count.head(10))
```

```python
#Question 2 the 10 most common project themes

#Pivot df to count the most commong project themes
project_count2=df.pivot_table(index=['code','theme'],aggfunc='count').reset_index()
project_count2.columns=['code','theme','Count of Theme','Count of Theme2']
del project_count2['Count of Theme2']
project_count2.sort_values(by=['Count of Theme'], inplace=True,ascending=False)
print(project_count2.head(10))
```

```python
#Here we can see that some of the code column does not have a theme associated with it.
project_count2
```

```python
#Here we create a dataframe of code and name
mjtheme_dict=json_normalize(data2,'mjtheme_namecode')

mjtheme_dict = mjtheme_dict.sort_values(by=['code'])
#Create a filter to find rows that are not ''
filter1=mjtheme_dict['name']!=''
#Apply the filter
mjtheme_dict=mjtheme_dict.loc[filter1]
#Drop any duplicates of name and code but keep the first value
mjtheme_dict.drop_duplicates(keep='first',inplace=True)
mjtheme_dict

#Create a list of the code values
code_list=list(mjtheme_dict['code'])
code_list
#Create a list of the name values
name_list=list(mjtheme_dict['name'])
name_list
#Create a dictionary of the code and name
final_dict=dict(zip(code_list,name_list))
final_dict
```

```python
#Question 3 the project where we creat a new column with the missing Theme
#We make sure th code column is a string
project_count2['code']=project_count2['code'].astype(str)

#We use the map function to bring in the missing Theme values
project_count2['Theme']=project_count2['code'].map(final_dict)
#We name the columns
project_count2.columns=['code','theme','Count of Theme','Theme']
#We print out the dataframe to inspect
project_count2
```

```python
#Question 3 but this time we repivot the data so that all the Themes are correctly summed
project_count3=project_count2.pivot_table(index=['code','Theme'],values='Count of Theme',aggfunc='sum').reset_index()
project_count3.sort_values(by=['Count of Theme'], inplace=True,ascending=False)
#We print out the dataframe to see the more accurate code,theme and count of theme
project_count3
```
